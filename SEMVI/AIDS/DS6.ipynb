```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

loan_data = pd.read_csv('/content/home_loan_train.csv')
label_columns = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Dependents']
for col in label_columns:
    if col == 'Dependents':
        loan_data[col] = loan_data[col].str.replace('+', '').astype(str)
    loan_data[col] = LabelEncoder().fit_transform(loan_data[col].astype(str))
loan_data = loan_data.drop('Loan_ID', axis=1)
loan_data = loan_data.fillna(loan_data.mean(numeric_only=True))
X = loan_data.drop('Loan_Status', axis=1)
Y = LabelEncoder().fit_transform(loan_data['Loan_Status'].astype(str))
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
Naivebayes = GaussianNB()
Naivebayes.fit(X_train, y_train)
y_pred = Naivebayes.predict(X_test)
print(y_pred)
# Evaluate accuracy
acc = sklearn.metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

## S&P500 dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load the S&P 500 companies dataset
data = pd.read_csv("/content/sp500_companies.csv")

# Display column names to identify target and features
print("Columns:", data.columns)

# Assume we are classifying companies based on Sector
# Drop non-numeric columns except the target (we'll encode them)
target_col = 'Sector'  # Change this if your classification target is different

# Drop irrelevant columns (like Symbol, Name, etc.)
data = data.drop(['Symbol', 'Security', 'GICS Sub-Industry', 'Headquarters Location',
                  'Date first added', 'CIK', 'Founded'], axis=1, errors='ignore')

# Encode categorical columns and handle missing values properly
for col in data.columns:
    if data[col].dtype == 'object' and col != target_col: # Ensure target is not encoded
        # Impute missing values before encoding
        data[col] = data[col].fillna(data[col].mode()[0])
        # Perform label encoding
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col].astype(str))
    else:
        data[col] = data[col].fillna(data[col].mean())


# Separate features and target
X = data.drop(target_col, axis=1)
y = data[target_col]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Train Classifiers
# ---------------------

# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)

# SVM
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
```
